```python
X=cancer_df.drop('target',axis=1)
y=cancer_df['target']#seperating target and rest of data
```

```python
#data preprocessing
#SPLITTING
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=2020)
```

```python
print(X_test.iloc[:1].to_string())
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_sc=sc.fit_transform(X_train)
X_test_sc=sc.fit_transform(X_test)
pickle.dump(sc,open('scale.pkl','wb'))
```

```python
sc1=pickle.load(open('scale.pkl','rb'))
ty=sc1.transform(X_test.iloc[:1])
ty
```

```python
X_test_sc[0]
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
#support vector classifier
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,y_train)
y_pred_svc=svc.predict(X_test)
accuracy_score(y_test,y_pred_svc)
```

```python
#support vector classifier with scaling
from sklearn.svm import SVC
svc2=SVC()
svc2.fit(X_train_sc,y_train)
y_pred_svc2=svc2.predict(X_test_sc)
accuracy_score(y_test,y_pred_svc2),svc2.score(X_test_sc,y_test)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred_lr=lr.predict(X_test)
lr.score(X_test,y_test)
```

```python
#linear regression with fc
from sklearn.linear_model import LinearRegression
lr2=LinearRegression()
lr2.fit(X_train_sc,y_train)
y_pred_lr2=lr2.predict(X_test_sc)
lr2.score(X_test_sc,y_test)
```

```python
#KNN classifier
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
knc.fit(X_train,y_train)
y_pred_knc=knc.predict(X_test)
knc.score(X_test,y_test)
```

```python
#KNN classifier with sc
from sklearn.neighbors import KNeighborsClassifier
knc2=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
knc2.fit(X_train_sc,y_train)
y_pred_knc2=knc2.predict(X_test_sc)
knc2.score(X_test_sc,y_test)
```

```python
#naive bayes
from sklearn.naive_bayes import GaussianNB
nbgc=GaussianNB()
nbgc.fit(X_train,y_train)
y_pred_nbgc=nbgc.predict(X_test)
nbgc.score(X_test,y_test)
```

```python
#naive bayes with fc or sc
from sklearn.naive_bayes import GaussianNB
nbgc2=GaussianNB()
nbgc2.fit(X_train_sc,y_train)
y_pred_nbgc2=nbgc2.predict(X_test_sc)
nbgc2.score(X_test_sc,y_test)
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(criterion='entropy')
dtc.fit(X_train,y_train)
y_pred_dtc=dtc.predict(X_test)
dtc.score(X_test,y_test)
```

```python
#decision tree with fc
from sklearn.tree import DecisionTreeClassifier
dtc2=DecisionTreeClassifier(criterion='entropy')
dtc2.fit(X_train_sc,y_train)
y_pred_dtc2=dtc2.predict(X_test_sc)
dtc2.score(X_test_sc,y_test)
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=20,criterion='entropy')
rfc.fit(X_train,y_train)
y_pred_rfc=rfc.predict(X_test)
rfc.score(X_test,y_test)
```

```python
#random forest with fc
from sklearn.ensemble import RandomForestClassifier
rfc2=RandomForestClassifier(n_estimators=20,criterion='entropy')
rfc2.fit(X_train_sc,y_train)
y_pred_rfc2=rfc2.predict(X_test_sc)
rfc2.score(X_test_sc,y_test)
```

```python
#confussion matrix as sv2 predict with max accuracy
cm=confusion_matrix(y_test,y_pred_svc2)
sns.heatmap(cm,annot=True)
plt.show()
```

```python
#classification report
print(classification_report(y_test,y_pred_svc2))
```

```python
# Cross validation
from sklearn.model_selection import cross_val_score
cross_validation = cross_val_score(estimator = svc2, X = X_train_sc, y = y_train, cv = 10)
print("Cross validation of support vector machine model = ",cross_validation)
print("Cross validation of support vector machine model (in mean) = ",cross_validation.mean())
```
